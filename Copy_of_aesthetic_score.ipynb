```python
# @title
#这个是可用的，对单独一个文件，不要动
# -*- coding: utf-8 -*-
# 安装依赖（如果尚未安装）
#!pip install torch torchvision scenedetect moviepy scikit-image

# 如果未下载 VBench，请取消下面注释：
# !git clone https://github.com/Vchitect/VBench.git

import os
import sys
import subprocess
import json
import shutil
import numpy as np
import cv2
from PIL import Image
from moviepy.editor import VideoFileClip
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
from skimage.metrics import structural_similarity as ssim

# ========= 环境设置 =========
vbench_dir = "VBench"  # 请确保该目录存在且已下载 VBench 项目
os.environ["VBENCH_DIR"] = vbench_dir
sys.path.append(vbench_dir)

# ========= VBench 多维度评分函数 =========
def evaluate_vbench_multi_dimensions(video_path, dimensions, vbench_dir="VBench", temp_dir=None):
    """
    使用 VBench 对单个视频进行评测，支持多个评分维度，
    返回这些评分的简单平均值作为该视频的得分。

    参数：
      - video_path: 待评测视频路径
      - dimensions: 评分维度列表，例如 ["aesthetic_quality", "spatial_relationship"]
      - vbench_dir: VBench 仓库路径（默认 "VBench"）
      - temp_dir: 临时工作目录（如果为 None，则根据视频文件名在 vbench_temp 下创建独立目录）

    返回：
      平均得分（浮点数），评分取自输出 JSON 中对应维度的第一个数值的平均值。

    同时，将 temp_dir 下的所有评测结果保存至对应视频输出目录的 vbench_results 子目录中。
    """
    video_name = os.path.basename(video_path)
    # 如果未指定临时目录，则使用 "vbench_temp/{video_name}"
    if temp_dir is None:
        temp_dir = os.path.join("vbench_temp", os.path.splitext(video_name)[0])
    os.makedirs(temp_dir, exist_ok=True)

    # 复制视频到 temp_dir/input 下
    staged_dir = os.path.join(temp_dir, "input")
    os.makedirs(staged_dir, exist_ok=True)
    shutil.copy(video_path, os.path.join(staged_dir, video_name))

    # 指定输出路径（结果可能为目录或文件）
    output_path = os.path.join(temp_dir, f"{video_name}.json")
    cmd = [
        "python", os.path.join(vbench_dir, "evaluate.py"),
        "--videos_path", staged_dir,
        "--dimension"
    ] + dimensions + [
        "--output_path", output_path,
        "--mode", "custom_input"
    ]
    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)
    except subprocess.CalledProcessError as e:
        print("🔥 VBench 出错！stderr:")
        print(e.stderr)
        print("🔥 stdout:")
        print(e.stdout)
        raise e

    # 如果 output_path 是目录，则从中查找第一个 JSON 文件
    if os.path.isdir(output_path):
        json_files = [os.path.join(output_path, fn) for fn in os.listdir(output_path) if fn.endswith('.json')]
        if not json_files:
            raise RuntimeError(f"在目录 {output_path} 中未找到 JSON 文件")
        result_file = json_files[0]
    else:
        result_file = output_path

    print("输出的 JSON 结果文件:", result_file)
    with open(result_file, "r") as f:
        data = json.load(f)
    print("JSON 内容:")
    print(json.dumps(data, indent=2))

    # 如果数据为列表，则取第一个元素，否则直接解析
    if isinstance(data, list) and len(data) > 0:
        data = data[0]

    # 尝试获取各维度得分
    scores = []
    for dim in dimensions:
        if dim in data:
            scores.append(float(data[dim][0]))
        else:
            print(f"⚠️ 结果 JSON 文件中未包含维度: {dim}")
    if not scores:
        raise ValueError(f"结果 JSON 中未找到任何评分，传入维度: {dimensions}")
    avg_score = sum(scores) / len(scores)

    # 将 temp_dir 下的所有评测结果移动到最终视频片段所在目录下的 vbench_results 子目录
    final_results_dir = os.path.join(os.path.dirname(video_path), "vbench_results", os.path.splitext(video_name)[0])
    os.makedirs(final_results_dir, exist_ok=True)
    # 将 temp_dir 内所有内容复制到 final_results_dir 下
    for item in os.listdir(temp_dir):
        s = os.path.join(temp_dir, item)
        d = os.path.join(final_results_dir, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)
    # 可选：清除 temp_dir
    shutil.rmtree(temp_dir)

    return avg_score

# ========= 计算场景相似度 =========
def compute_scene_similarity(segment_paths, num_samples=3):
    """
    计算所有视频片段之间的最小 SSIM 值，
    并返回该最小 SSIM 值。
    """
    segment_frames = []
    for path in segment_paths:
        clip = VideoFileClip(path)
        duration = clip.duration
        times = np.linspace(0, duration, num=num_samples+2)[1:-1]
        frames = []
        for t in times:
            try:
                frame = clip.get_frame(t)
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                frames.append(gray)
            except Exception as e:
                print(f"Error sampling frame at t={t}: {e}")
        segment_frames.append(frames)
        clip.close()

    min_ssims = []
    for i in range(len(segment_frames)):
        for j in range(i+1, len(segment_frames)):
            ssims = []
            for f1 in segment_frames[i]:
                for f2 in segment_frames[j]:
                    try:
                        s_val = ssim(f1, f2)
                        ssims.append(s_val)
                    except Exception as e:
                        print(f"Error computing SSIM: {e}")
            if ssims:
                min_ssims.append(min(ssims))
    return min(min_ssims) if min_ssims else 0.0

# ========= 场景检测 =========
def detect_scenes(video_path, threshold=10.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    return [(scene[0].get_seconds(), scene[1].get_seconds()) for scene in scene_list]

# ========= 视频切割 =========
def cut_video_by_scenes(video_path, scenes, output_dir='video_segments'):
    os.makedirs(output_dir, exist_ok=True)
    clip = VideoFileClip(video_path)
    segments = []
    if not scenes:
        output_path = os.path.join(output_dir, 'segment_0.mp4')
        clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
        segments.append((output_path, 0, clip.duration))
    else:
        for i, (start, end) in enumerate(scenes):
            output_path = os.path.join(output_dir, f'segment_{i}.mp4')
            subclip = clip.subclip(start, end)
            subclip.write_videofile(output_path, codec='libx264', audio_codec='aac')
            segments.append((output_path, start, end))
    clip.close()
    return segments

# ========= 主流程 =========
def main():
    input_video_path = '/content/my_videos_input/easyanimate-videos/00000001.mp4'  # 请上传视频
    base_name = os.path.splitext(os.path.basename(input_video_path))[0]
    output_dir = os.path.join('video_segments', base_name)
    os.makedirs(output_dir, exist_ok=True)

    # 设定场景切割阈值，并归一化（阈值除以100）
    threshold_raw = 10.0
    threshold_score_norm = min(max(threshold_raw / 100.0, 0), 1)

    # Step 1: 检测场景转场
    scenes = detect_scenes(input_video_path, threshold=threshold_raw)

    # Step 2: 切割视频为片段
    segments = cut_video_by_scenes(input_video_path, scenes, output_dir=output_dir)

    # Step 3: 使用 VBench 多维度评分（例如 ["aesthetic_quality", "spatial_relationship"]）
    dimensions = ["aesthetic_quality", "overall_consistency", "temporal_style", "background_consistency" ,"imaging_quality" ,"motion_smoothness" , "dynamic_degree", "subject_consistency" ,"temporal_flickering"]  # 示例维度，可根据需要修改
    segment_scores = []
    for seg_path, start, end in segments:
        try:
            # 使用基于每个视频片段的独立临时目录进行评测
            temp_dir = os.path.join("vbench_temp", os.path.splitext(os.path.basename(seg_path))[0])
            score = evaluate_vbench_multi_dimensions(seg_path, dimensions, vbench_dir=vbench_dir, temp_dir=temp_dir)
            segment_scores.append((seg_path, start, end, score))
            print(f"片段: {seg_path} (时间: {start:.1f}s - {end:.1f}s) 得分: {score:.2f}")
        except Exception as e:
            print(f"❌ VBench评分失败: {seg_path} | 错误: {e}")
            segment_scores.append((seg_path, start, end, 0.0))

    # Step 4: 归一化评分：将评分直接除以 10（映射到 [0,1]）
    normalized_scores = [
        (seg_path, start, end, raw_score, min(max(raw_score / 10.0, 0.0), 1.0))
        for seg_path, start, end, raw_score in segment_scores
    ]

    # Step 5: 计算各片段之间的场景相似度（使用 SSIM 最小值），归一化为“不相似度”(1 - sim)
    similarity_score = compute_scene_similarity([s[0] for s in segment_scores]) if len(segment_scores) > 1 else 1.0
    similarity_norm = 1 - similarity_score

    # Step 6: 综合评分计算
    score_highest_norm = max([s[4] for s in normalized_scores])
    score_lowest_norm = min([s[4] for s in normalized_scores])
    w1, w2, w3, w4 = 0.3, 0.3, 0.1, 0.3
    final_score = (
        w1 * score_highest_norm +
        w2 * score_lowest_norm +
        w3 * threshold_score_norm +
        w4 * similarity_norm
    )

    # Step 7: 将所有详细评分信息写入 scores.txt 文件，并将 vbench_temp 的内容移动到最终输出目录中
    scores_file_path = os.path.join(output_dir, 'scores.txt')
    with open(scores_file_path, 'w') as f:
        f.write(f"# 输入视频: {input_video_path}\n")
        f.write(f"# 检测到 {len(segments)} 个片段\n\n")
        f.write("# 各片段评分（原始得分及归一化得分）：\n")
        for seg_path, start, end, raw_score, norm_score in normalized_scores:
            f.write(f"{os.path.basename(seg_path)} | 时间: {start:.1f}-{end:.1f}s | 原始得分: {raw_score:.2f} | 归一化得分: {norm_score:.4f}\n")
        f.write(f"\n阈值: {threshold_raw:.2f} → 归一化阈值: {threshold_score_norm:.4f}\n")
        f.write(f"场景相似度（最小 SSIM）: {similarity_score:.4f} → 不相似度: {similarity_norm:.4f}\n\n")
        f.write("# 权重参数:\n")
        f.write(f"w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}\n\n")
        f.write("# 综合评分计算公式:\n")
        f.write(f"final_score = {w1} * {score_highest_norm:.4f} + {w2} * {score_lowest_norm:.4f} + {w3} * {threshold_score_norm:.4f} + {w4} * {similarity_norm:.4f}\n")
        f.write(f"\n✅ 综合美学评分（归一化）: {final_score:.4f}\n")

    # 将 vbench_temp 下的内容复制到最终输出目录下的 "vbench_results" 子文件夹中
    final_results_dir = os.path.join(output_dir, "vbench_results")
    os.makedirs(final_results_dir, exist_ok=True)
    if os.path.exists("vbench_temp"):
        for item in os.listdir("vbench_temp"):
            s = os.path.join("vbench_temp", item)
            d = os.path.join(final_results_dir, item)
            if os.path.isdir(s):
                shutil.copytree(s, d, dirs_exist_ok=True)
            else:
                shutil.copy2(s, d)
        # 可选：清空 vbench_temp 文件夹
        shutil.rmtree("vbench_temp")

    print(f"\n✅ 综合评分: {final_score:.4f}")
    print(f"✅ 详细评分结果已保存至: {scores_file_path}")
    print(f"✅ VBench 结果文件已移动至: {final_results_dir}")

if __name__ == '__main__':
    main()
```

```python
# @title
# -*- coding: utf-8 -*-
#对文件夹进行批量处理，打分

# -*- coding: utf-8 -*-
#对文件夹进行批量处理，打分

import os
import sys
import subprocess
import json
import shutil
import numpy as np
import cv2
import csv
from PIL import Image
from moviepy.editor import VideoFileClip
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
from skimage.metrics import structural_similarity as ssim

# ========= 环境设置 =========
vbench_dir = "VBench"  # 请确保该目录存在且已下载 VBench 项目
os.environ["VBENCH_DIR"] = vbench_dir
sys.path.append(vbench_dir)

# ========= VBench 多维度评分函数 =========
def evaluate_vbench_multi_dimensions(video_path, dimensions, vbench_dir="VBench", temp_dir=None):
    """
    使用 VBench 对单个视频进行评测，支持多个评分维度，
    返回这些评分的简单平均值作为该视频的得分。

    参数：
      - video_path: 待评测视频路径
      - dimensions: 评分维度列表，例如 ["aesthetic_quality", "spatial_relationship"]
      - vbench_dir: VBench 仓库路径（默认 "VBench"）
      - temp_dir: 临时工作目录（如果为 None，则根据视频文件名在对应视频切割目录下创建，
                  格式为 video_segments/<视频基本名>/<视频文件名>_temp ）

    返回：
      平均得分（浮点数），评分取自输出 JSON 中对应维度的第一个数值的平均值。

    同时，将 temp_dir 下的所有评测结果保存至对应视频输出目录的 vbench_results 子目录中。
    """
    video_name = os.path.basename(video_path)
    if temp_dir is None:
        base_name = os.path.splitext(video_name)[0]
        temp_dir = os.path.join("video_segments", base_name, base_name + "_temp")
    os.makedirs(temp_dir, exist_ok=True)

    # 复制视频到 temp_dir/input 下
    staged_dir = os.path.join(temp_dir, "input")
    os.makedirs(staged_dir, exist_ok=True)
    shutil.copy(video_path, os.path.join(staged_dir, video_name))

    # 指定输出路径（结果可能为目录或文件）
    output_path = os.path.join(temp_dir, f"{video_name}.json")
    cmd = [
        "python", os.path.join(vbench_dir, "evaluate.py"),
        "--videos_path", staged_dir,
        "--dimension"
    ] + dimensions + [
        "--output_path", output_path,
        "--mode", "custom_input"
    ]
    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)
    except subprocess.CalledProcessError as e:
        print("🔥 VBench 出错！stderr:")
        print(e.stderr)
        print("🔥 stdout:")
        print(e.stdout)
        raise e

    # 修改部分：读取所有 JSON 文件中评分
    all_scores = []
    if os.path.isdir(output_path):
        json_files = [os.path.join(output_path, fn) for fn in os.listdir(output_path) if fn.endswith('.json')]
        if not json_files:
            raise RuntimeError(f"在目录 {output_path} 中未找到 JSON 文件")
        for jf in json_files:
            with open(jf, "r") as f:
                data = json.load(f)
            if isinstance(data, list) and len(data) > 0:
                data = data[0]
            file_scores = []
            for dim in dimensions:
                if dim in data:
                    file_scores.append(float(data[dim][0]))
                else:
                    print(f"⚠️ 文件 {jf} 中未包含维度: {dim}")
            if file_scores:
                all_scores.append(sum(file_scores) / len(file_scores))
    else:
        with open(output_path, "r") as f:
            data = json.load(f)
        if isinstance(data, list) and len(data) > 0:
            data = data[0]
        file_scores = []
        for dim in dimensions:
            if dim in data:
                file_scores.append(float(data[dim][0]))
            else:
                print(f"⚠️ 文件 {output_path} 中未包含维度: {dim}")
        if file_scores:
            all_scores.append(sum(file_scores) / len(file_scores))

    if not all_scores:
        raise ValueError(f"结果 JSON 中未找到任何评分，传入维度: {dimensions}")
    avg_score = sum(all_scores) / len(all_scores)

    # 将 temp_dir 下的所有评测结果移动到最终视频片段所在目录下的 vbench_results 子目录
    final_results_dir = os.path.join(os.path.dirname(video_path), "vbench_results", os.path.splitext(video_name)[0])
    os.makedirs(final_results_dir, exist_ok=True)
    for item in os.listdir(temp_dir):
        s = os.path.join(temp_dir, item)
        d = os.path.join(final_results_dir, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)
    shutil.rmtree(temp_dir)

    return avg_score

# ========= 计算场景相似度 =========
def compute_scene_similarity(segment_paths, num_samples=3):
    """
    计算所有视频片段之间的最小 SSIM 值，
    并返回该最小 SSIM 值。
    """
    segment_frames = []
    for path in segment_paths:
        clip = VideoFileClip(path)
        duration = clip.duration
        times = np.linspace(0, duration, num=num_samples+2)[1:-1]
        frames = []
        for t in times:
            try:
                frame = clip.get_frame(t)
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                frames.append(gray)
            except Exception as e:
                print(f"Error sampling frame at t={t}: {e}")
        segment_frames.append(frames)
        clip.close()

    min_ssims = []
    for i in range(len(segment_frames)):
        for j in range(i+1, len(segment_frames)):
            ssims = []
            for f1 in segment_frames[i]:
                for f2 in segment_frames[j]:
                    try:
                        s_val = ssim(f1, f2)
                        ssims.append(s_val)
                    except Exception as e:
                        print(f"Error computing SSIM: {e}")
            if ssims:
                min_ssims.append(min(ssims))
    return min(min_ssims) if min_ssims else 0.0

# ========= 场景检测 =========
def detect_scenes(video_path, threshold=10.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    return [(scene[0].get_seconds(), scene[1].get_seconds()) for scene in scene_list]

# ========= 视频切割 =========
def cut_video_by_scenes(video_path, scenes, output_dir='/content/testvideo_score/easyanimate/video_segments'):
    os.makedirs(output_dir, exist_ok=True)
    clip = VideoFileClip(video_path)
    segments = []
    if not scenes:
        output_path = os.path.join(output_dir, 'segment_0.mp4')
        clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
        segments.append((output_path, 0, clip.duration))
    else:
        for i, (start, end) in enumerate(scenes):
            output_path = os.path.join(output_dir, f'segment_{i}.mp4')
            subclip = clip.subclip(start, end)
            subclip.write_videofile(output_path, codec='libx264', audio_codec='aac')
            segments.append((output_path, start, end))
    clip.close()
    return segments

# ========= 文件夹视频批量处理 =========
def process_folder(input_folder, dimensions, output_csv):
    results = []  # 用于存储每个视频的评分信息
    # 遍历输入文件夹中的所有 mp4 文件
    for file_name in os.listdir(input_folder):
        if file_name.lower().endswith(".mp4"):
            video_path = os.path.join(input_folder, file_name)
            print(f"处理视频: {video_path}")

            base_name = os.path.splitext(file_name)[0]
            output_dir = os.path.join("video_segments", base_name)
            os.makedirs(output_dir, exist_ok=True)

            # Step 1: 检测场景转场
            scenes = detect_scenes(video_path, threshold=10.0)
            # Step 2: 切割视频为片段
            segments = cut_video_by_scenes(video_path, scenes, output_dir=output_dir)

            # Step 3: 针对每个片段评分
            segment_scores = []
            for seg_path, start, end in segments:
                try:
                    # 临时目录为 video_segments/<视频名>/<片段文件名>_temp
                    temp_dir = os.path.join(output_dir, os.path.splitext(os.path.basename(seg_path))[0] + "_temp")
                    score = evaluate_vbench_multi_dimensions(seg_path, dimensions, vbench_dir=vbench_dir, temp_dir=temp_dir)
                    segment_scores.append((seg_path, start, end, score))
                    print(f"片段: {seg_path} (时间: {start:.1f}s - {end:.1f}s) 得分: {score:.2f}")
                except Exception as e:
                    print(f"❌ VBench评分失败: {seg_path} | 错误: {e}")
                    segment_scores.append((seg_path, start, end, 0.0))

            if segment_scores:
                avg_raw = sum([s[3] for s in segment_scores]) / len(segment_scores)
            else:
                avg_raw = 0.0

            # Step 4: 归一化评分：将原始得分除以 1（直接映射到 [0,1]）
            normalized_scores = [
                (seg_path, start, end, raw_score, min(max(raw_score / 1.0, 0.0), 1.0))
                for seg_path, start, end, raw_score in segment_scores
            ]
            norm_raw = min(max(avg_raw / 1.0, 0.0), 1.0)

            # Step 5: 计算各片段之间的场景相似度（使用 SSIM 最小值），归一化为“不相似度”(1 - sim)
            if len(segments) > 1:
                similarity_score = compute_scene_similarity([s[0] for s in segment_scores])
            else:
                similarity_score = 1.0
            similarity_norm = 1 - similarity_score

            # Step 6: 综合评分计算（权重参数与原代码一致）
            score_highest_norm = max([s[4] for s in normalized_scores]) if normalized_scores else 0.0
            score_lowest_norm = min([s[4] for s in normalized_scores]) if normalized_scores else 0.0
            threshold_raw = 10.0
            threshold_score_norm = min(max(threshold_raw / 100.0, 0.0), 1.0)
            w1, w2, w3, w4 = 0.3, 0.3, 0.1, 0.3
            final_score = (
                w1 * score_highest_norm +
                w2 * score_lowest_norm +
                w3 * threshold_score_norm +
                w4 * similarity_norm
            )

            # Step 7: 输出每个视频详细评分信息至 scores.txt，确保片段序号正确
            scores_file_path = os.path.join(output_dir, 'scores.txt')
            with open(scores_file_path, 'w') as f:
                f.write(f"# 输入视频: {video_path}\n")
                f.write(f"# 检测到 {len(segments)} 个片段\n\n")
                f.write("# 各片段评分（按序号、原始得分及归一化得分）：\n")
                for idx, (seg_path, start, end, raw_score, norm_score) in enumerate(normalized_scores):
                    f.write(f"Segment {idx}: {os.path.basename(seg_path)} | 时间: {start:.1f}-{end:.1f}s | 原始得分: {raw_score:.2f} | 归一化得分: {norm_score:.4f}\n")
                f.write(f"\n阈值: {threshold_raw:.2f} → 归一化阈值: {threshold_score_norm:.4f}\n")
                f.write(f"场景相似度（最小 SSIM）: {similarity_score:.4f} → 不相似度: {similarity_norm:.4f}\n\n")
                f.write("# 权重参数:\n")
                f.write(f"w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}\n\n")
                f.write("# 综合评分计算公式:\n")
                f.write(f"final_score = {w1} * {score_highest_norm:.4f} + {w2} * {score_lowest_norm:.4f} + {w3} * {threshold_score_norm:.4f} + {w4} * {similarity_norm:.4f}\n")
                f.write(f"\n✅ 综合美学评分（归一化）: {final_score:.4f}\n")
            print(f"详细评分结果已保存至: {scores_file_path}")

            results.append({
                "file_name": file_name,
                "original_score": avg_raw,
                "normalized_score": norm_raw,
                "similarity_norm": similarity_norm,
                "final_score": final_score
            })

    # Step 8: 将所有视频的评分信息写入 CSV 文件
    with open(output_csv, "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["file_name", "original_score", "normalized_score", "similarity_norm", "final_score"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for row in results:
            writer.writerow(row)
    print(f"\n✅ 评分 CSV 文件已保存至: {output_csv}")

def main():
    # 修改此处为包含待评分视频的文件夹路径，例如 "input_videos"
    input_folder = '/content/my_videos_input/easyanimate-videos/workspace/EasyAnimate/samples/easyanimate-videos'
    dimensions = ["aesthetic_quality", "overall_consistency", "temporal_style", "background_consistency", "imaging_quality", "motion_smoothness", "dynamic_degree", "subject_consistency", "temporal_flickering"]
    output_csv = "video_scores.csv"

    process_folder(input_folder, dimensions, output_csv)

if __name__ == '__main__':
    main()
```

```python
# -*- coding: utf-8 -*-
# Colab-ready 脚本：批量视频评分（带进度条、实时 CSV 更新、记录详细维度评分）

# 安装依赖
!pip install tqdm moviepy scenedetect scikit-image

import os
import sys
import subprocess
import json
import shutil
import numpy as np
import cv2
import csv
from tqdm import tqdm
from PIL import Image
from moviepy.editor import VideoFileClip
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
from skimage.metrics import structural_similarity as ssim

# ========= 环境设置 =========
vbench_dir = "VBench"  # 请确保该目录存在且已下载 VBench 项目
os.environ["VBENCH_DIR"] = vbench_dir
sys.path.append(vbench_dir)

def evaluate_vbench_multi_dimensions(video_path, dimensions, vbench_dir="VBench", temp_dir=None):
    """
    使用 VBench 对单个视频进行评测，支持多个评分维度，
    返回平均得分以及一个包含各维度分数的字典（所有 JSON 文件中各维度的平均分）。

    参数：
      - video_path: 待评测视频路径
      - dimensions: 评分维度列表，例如 ["aesthetic_quality", "spatial_relationship"]
      - vbench_dir: VBench 仓库路径（默认 "VBench"）
      - temp_dir: 临时工作目录（如果为 None，则根据视频文件名在对应视频切割目录下创建，
                  格式为 video_segments/<视频基本名>/<视频文件名>_temp ）
    """
    video_name = os.path.basename(video_path)
    if temp_dir is None:
        base_name = os.path.splitext(video_name)[0]
        temp_dir = os.path.join("video_segments", base_name, base_name + "_temp")
    os.makedirs(temp_dir, exist_ok=True)

    # 复制视频到 temp_dir/input 下
    staged_dir = os.path.join(temp_dir, "input")
    os.makedirs(staged_dir, exist_ok=True)
    shutil.copy(video_path, os.path.join(staged_dir, video_name))

    # 指定输出路径（结果可能为目录或文件）
    output_path = os.path.join(temp_dir, f"{video_name}.json")
    cmd = [
        "python", os.path.join(vbench_dir, "evaluate.py"),
        "--videos_path", staged_dir,
        "--dimension"
    ] + dimensions + [
        "--output_path", output_path,
        "--mode", "custom_input"
    ]
    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)
    except subprocess.CalledProcessError as e:
        print("🔥 VBench 出错！stderr:")
        print(e.stderr)
        print("🔥 stdout:")
        print(e.stdout)
        raise e

    # 读取所有 JSON 文件中的评分
    all_scores = []
    all_dimension_scores = {dim: 0.0 for dim in dimensions}
    if os.path.isdir(output_path):
        json_files = [os.path.join(output_path, fn) for fn in os.listdir(output_path) if fn.endswith('.json')]
        if not json_files:
            raise RuntimeError(f"在目录 {output_path} 中未找到 JSON 文件")
    else:
        json_files = [output_path]

    for jf in json_files:
        with open(jf, "r") as f:
            data = json.load(f)
        if isinstance(data, list) and len(data) > 0:
            data = data[0]
        file_scores = []
        for dim in dimensions:
            if dim in data:
                score = float(data[dim][0])
                file_scores.append(score)
                all_dimension_scores[dim] += score
            else:
                print(f"⚠️ 文件 {jf} 中未包含维度: {dim}")
        if file_scores:
            all_scores.append(sum(file_scores) / len(file_scores))

    if not all_scores:
        raise ValueError(f"结果 JSON 中未找到任何评分，传入维度: {dimensions}")
    # 计算各维度平均分
    for dim in dimensions:
        all_dimension_scores[dim] /= len(json_files)
    avg_score = sum(all_scores) / len(all_scores)

    # 将 temp_dir 下的所有评测结果移动到最终视频片段所在目录下的 vbench_results 子目录
    final_results_dir = os.path.join(os.path.dirname(video_path), "vbench_results", os.path.splitext(video_name)[0])
    os.makedirs(final_results_dir, exist_ok=True)
    for item in os.listdir(temp_dir):
        s = os.path.join(temp_dir, item)
        d = os.path.join(final_results_dir, item)
        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)
    shutil.rmtree(temp_dir)

    return avg_score, all_dimension_scores

def compute_scene_similarity(segment_paths, num_samples=3):
    """
    计算所有视频片段之间的最小 SSIM 值，并返回该值。
    """
    segment_frames = []
    for path in segment_paths:
        clip = VideoFileClip(path)
        duration = clip.duration
        times = np.linspace(0, duration, num=num_samples+2)[1:-1]
        frames = []
        for t in times:
            try:
                frame = clip.get_frame(t)
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                frames.append(gray)
            except Exception as e:
                print(f"Frame error at {t}: {e}")
        segment_frames.append(frames)
        clip.close()
    min_ssims = []
    for i in range(len(segment_frames)):
        for j in range(i+1, len(segment_frames)):
            ssims = []
            for f1 in segment_frames[i]:
                for f2 in segment_frames[j]:
                    try:
                        ssims.append(ssim(f1, f2))
                    except Exception as e:
                        print(f"SSIM error: {e}")
            if ssims:
                min_ssims.append(min(ssims))
    return min(min_ssims) if min_ssims else 0.0

def detect_scenes(video_path, threshold=10.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    return [(scene[0].get_seconds(), scene[1].get_seconds()) for scene in scene_list]

def cut_video_by_scenes(video_path, scenes, output_dir='/video_segments'):
    os.makedirs(output_dir, exist_ok=True)
    clip = VideoFileClip(video_path)
    segments = []
    if not scenes:
        output_path = os.path.join(output_dir, 'segment_0.mp4')
        clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
        segments.append((output_path, 0, clip.duration))
    else:
        for i, (start, end) in enumerate(scenes):
            output_path = os.path.join(output_dir, f'segment_{i}.mp4')
            subclip = clip.subclip(start, end)
            subclip.write_videofile(output_path, codec='libx264', audio_codec='aac')
            segments.append((output_path, start, end))
    clip.close()
    return segments

def process_folder(input_folder, dimensions, output_csv):
    header_written = False
    for file_name in tqdm(os.listdir(input_folder), desc="处理视频中"):
        if not file_name.lower().endswith(".mp4"):
            continue
        video_path = os.path.join(input_folder, file_name)
        base_name = os.path.splitext(file_name)[0]
        output_dir = os.path.join("video_segments", base_name)
        os.makedirs(output_dir, exist_ok=True)
        scenes = detect_scenes(video_path, threshold=10.0)
        segments = cut_video_by_scenes(video_path, scenes, output_dir=output_dir)
        segment_scores = []
        for seg_path, start, end in segments:
            try:
                temp_dir = os.path.join(output_dir, os.path.splitext(os.path.basename(seg_path))[0] + "_temp")
                score, dim_scores = evaluate_vbench_multi_dimensions(seg_path, dimensions, vbench_dir=vbench_dir, temp_dir=temp_dir)
                segment_scores.append((seg_path, start, end, score, dim_scores))
                print(f"片段: {seg_path} (时间: {start:.1f}-{end:.1f}s) 得分: {score:.2f}")
            except Exception as e:
                print(f"评分失败: {seg_path}, 错误: {e}")
                segment_scores.append((seg_path, start, end, 0.0, {dim: 0.0 for dim in dimensions}))
        if segment_scores:
            avg_score = np.mean([s[3] for s in segment_scores])
        else:
            avg_score = 0.0
        norm_score = min(max(avg_score / 1.0, 0.0), 1.0)
        if len(segments) > 1:
            similarity_score = compute_scene_similarity([s[0] for s in segment_scores])
        else:
            similarity_score = 1.0
        similarity_norm = 1 - similarity_score
        norms = [min(max(s[3] / 1.0, 0.0), 1.0) for s in segment_scores]
        score_highest = max(norms) if norms else 0.0
        score_lowest = min(norms) if norms else 0.0
        threshold_score_norm = min(max(10.0 / 100.0, 0.0), 1.0)
        w1, w2, w3, w4 = 0.3, 0.3, 0.1, 0.3
        final_score = w1 * score_highest + w2 * score_lowest + w3 * threshold_score_norm + w4 * similarity_norm
        merged_scores = {dim: 0.0 for dim in dimensions}
        for _, _, _, _, d in segment_scores:
            for k in d:
                merged_scores[k] += d[k]
        if segment_scores:
            for k in merged_scores:
                merged_scores[k] = round(merged_scores[k] / len(segment_scores), 4)
        row = {
            "file_name": file_name,
            "original_score": round(avg_score, 4),
            "normalized_score": round(norm_score, 4),
            "similarity_norm": round(similarity_norm, 4),
            "final_score": round(final_score, 4)
        }
        row.update(merged_scores)
        with open(output_csv, "a", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=row.keys())
            if not header_written:
                writer.writeheader()
                header_written = True
            writer.writerow(row)
    print(f"\nCSV 文件已保存至: {output_csv}")

def main():
    input_folder = "/content/my_videos_input/train_1_3_nomotion_fps18_after_post_11520"  # 待处理视频文件夹路径
    dimensions = ["aesthetic_quality", "overall_consistency", "temporal_style", "background_consistency",
                  "imaging_quality", "motion_smoothness", "dynamic_degree", "subject_consistency", "temporal_flickering"]
    output_csv = "/content/video_scores.csv"
    process_folder(input_folder, dimensions, output_csv)

if __name__ == "__main__":
    main()

```

```python
import torch
from PIL import Image
from aesthetic_predictor_v2_5 import convert_v2_5_from_siglip

# 设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 加载模型和预处理器
model, preprocessor = convert_v2_5_from_siglip(
    low_cpu_mem_usage=True,
    trust_remote_code=True,
)
model = model.to(torch.bfloat16).to(device)
model.eval()
```

```python
import os
import numpy as np
from moviepy.editor import VideoFileClip
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
from skimage.metrics import structural_similarity as ssim
import cv2

def score_frame(frame):
    image = Image.fromarray(frame.astype('uint8'), mode='RGB')
    pixel_values = preprocessor(images=image, return_tensors="pt").pixel_values.to(torch.bfloat16).to(device)
    with torch.inference_mode():
        score = model(pixel_values).logits.squeeze().float().cpu().numpy()
    return score

def detect_scenes(video_path, threshold=30.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    return [(scene[0].get_seconds(), scene[1].get_seconds()) for scene in scene_list]

def cut_video_by_scenes(video_path, scenes, output_dir='video_segments'):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    clip = VideoFileClip(video_path)
    segments = []
    if not scenes:
        output_path = os.path.join(output_dir, 'segment_0.mp4')
        clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
        segments.append((output_path, 0, clip.duration))
    else:
        for i, (start, end) in enumerate(scenes):
            output_path = os.path.join(output_dir, f'segment_{i}.mp4')
            subclip = clip.subclip(start, end)
            subclip.write_videofile(output_path, codec='libx264', audio_codec='aac')
            segments.append((output_path, start, end))
    clip.close()
    return segments

def aesthetic_score(video_path, sample_interval=1.0):
    clip = VideoFileClip(video_path)
    duration = clip.duration
    scores = []
    t = 0.0
    while t < duration:
        try:
            frame = clip.get_frame(t)
            s = score_frame(frame)
            scores.append(s)
        except Exception as e:
            print(f"采样 t={t:.1f} 秒时出错: {e}")
        t += sample_interval
    clip.close()
    return np.mean(scores) if scores else 0

def compute_scene_similarity(segment_paths, num_samples=3):
    segment_frames = []
    for path in segment_paths:
        clip = VideoFileClip(path)
        duration = clip.duration
        times = np.linspace(0, duration, num=num_samples+2)[1:-1]
        frames = []
        for t in times:
            try:
                frame = clip.get_frame(t)
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                frames.append(gray)
            except Exception as e:
                print(f"采样 t={t:.1f} 秒时出错: {e}")
        segment_frames.append(frames)
        clip.close()

    min_ssims = []
    for i in range(len(segment_frames)):
        for j in range(i+1, len(segment_frames)):
            ssims = []
            for f1 in segment_frames[i]:
                for f2 in segment_frames[j]:
                    try:
                        s = ssim(f1, f2)
                        ssims.append(s)
                    except Exception as e:
                        print(f"计算 SSIM 时出错: {e}")
            if ssims:
                min_ssims.append(min(ssims))
    if not min_ssims:
        return 0.0
    min_ssim = min(min_ssims)
    return min_ssim
```

```python
import os
import subprocess
import numpy as np
import json
import cv2
from moviepy.editor import VideoFileClip
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
from skimage.metrics import structural_similarity as ssim

def detect_scenes(video_path, threshold=10.0):
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    return [(scene[0].get_seconds(), scene[1].get_seconds()) for scene in scene_list]

def cut_video_by_scenes(video_path, scenes, output_dir='video_segments'):
    os.makedirs(output_dir, exist_ok=True)
    clip = VideoFileClip(video_path)
    segments = []
    if not scenes:
        output_path = os.path.join(output_dir, 'segment_0.mp4')
        clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
        segments.append((output_path, 0, clip.duration))
    else:
        for i, (start, end) in enumerate(scenes):
            output_path = os.path.join(output_dir, f'segment_{i}.mp4')
            subclip = clip.subclip(start, end)
            subclip.write_videofile(output_path, codec='libx264', audio_codec='aac')
            segments.append((output_path, start, end))
    clip.close()
    return segments

def compute_scene_similarity(segment_paths, num_samples=3):
    segment_frames = []
    for path in segment_paths:
        clip = VideoFileClip(path)
        duration = clip.duration
        times = np.linspace(0, duration, num=num_samples+2)[1:-1]
        frames = []
        for t in times:
            try:
                frame = clip.get_frame(t)
                gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
                frames.append(gray)
            except:
                pass
        segment_frames.append(frames)
        clip.close()

    min_ssims = []
    for i in range(len(segment_frames)):
        for j in range(i+1, len(segment_frames)):
            ssims = []
            for f1 in segment_frames[i]:
                for f2 in segment_frames[j]:
                    try:
                        s = ssim(f1, f2)
                        ssims.append(s)
                    except:
                        pass
            if ssims:
                min_ssims.append(min(ssims))
    return min(min_ssims) if min_ssims else 0.0

def evaluate_vbench_single_video(video_path, vbench_dir="VBench", temp_dir="vbench_temp"):
    import shutil
    video_name = os.path.basename(video_path)
    staged_dir = os.path.join(temp_dir, "input")
    os.makedirs(staged_dir, exist_ok=True)
    shutil.copy(video_path, os.path.join(staged_dir, video_name))

    result_path = os.path.join(temp_dir, f"{video_name}.json")
    cmd = [
        "python", os.path.join(vbench_dir, "evaluate.py"),
        "--videos_path", staged_dir,
        "--dimension", "visual_quality",
        "--output_path", result_path,
        "--mode", "custom_input"
    ]

    try:
        result = subprocess.run(cmd, check=True, capture_output=True, text=True)
    except subprocess.CalledProcessError as e:
        print("🔥 VBench 出错！stderr:")
        print(e.stderr)
        print("🔥 stdout:")
        print(e.stdout)
        raise e

    if not os.path.exists(result_path):
        raise RuntimeError(f"未找到结果文件: {result_path}")

    with open(result_path, "r") as f:
        data = json.load(f)

    # 提取 visual_quality 得分
    if video_name not in data:
        raise ValueError(f"结果 JSON 中未找到视频条目: {video_name}")

    score = data[video_name].get("visual_quality", 0.0)
    return float(score)

```

```python
# /content/21768.mp4
# /content/EulerAncestralDiscrete_1_gs7.5_s100 (3).mp4
#/content/EulerAncestralDiscrete_6_gs7.5_s100 (1).mp4
# /content/21760.mp4
# 传统美学评分
#/content/EulerAncestralDiscrete_8_gs7.5_s100.mp4
import os

def main():
    input_video_path = '/content/21759.mp4'
    base_name = os.path.splitext(os.path.basename(input_video_path))[0]
    output_dir = os.path.join('video_segments', base_name)
    os.makedirs(output_dir, exist_ok=True)

    threshold_raw = 10.0
    threshold_score_norm = min(max(threshold_raw / 100.0, 0), 1)

    try:
        scenes = detect_scenes(input_video_path, threshold=threshold_raw)
        segments = cut_video_by_scenes(input_video_path, scenes, output_dir=output_dir)
    except Exception as e:
        print(f"❌ 场景检测或视频切割失败: {e}")
        return

    segment_scores = []
    for seg_path, start, end in segments:
        try:
            score = aesthetic_score(seg_path)
            segment_scores.append((seg_path, start, end, score))
        except Exception as e:
            print(f"❌ 美学评分失败（{seg_path}）: {e}")
            segment_scores.append((seg_path, start, end, 0.0))

    if not segment_scores:
        print("⚠️ 无有效片段评分，退出流程。")
        return

    # 美学评分归一化: score / 10.0
    normalized_scores = [
        (seg_path, start, end, raw_score, min(max(raw_score / 10.0, 0.0), 1.0))
        for seg_path, start, end, raw_score in segment_scores
    ]

    # 相似度计算
    try:
        similarity_score = compute_scene_similarity([s[0] for s in segment_scores]) if len(segment_scores) > 1 else 1.0
    except Exception as e:
        print(f"❌ 相似度计算失败: {e}")
        similarity_score = 1.0
    similarity_norm = 1 - similarity_score

    # 综合评分合成
    score_highest_norm = max([s[4] for s in normalized_scores])
    score_lowest_norm = min([s[4] for s in normalized_scores])

    w1, w2, w3, w4 = 0.3, 0.3, 0.1, 0.3
    final_score = (
        w1 * score_highest_norm +
        w2 * score_lowest_norm +
        w3 * threshold_score_norm +
        w4 * similarity_norm
    )

    # 保存评分
    scores_file_path = os.path.join(output_dir, 'scores.txt')
    with open(scores_file_path, 'w') as f:
        f.write(f"# 输入视频: {input_video_path}\n")
        f.write(f"# 检测到 {len(segments)} 个片段\n\n")

        f.write("# 各片段评分（含归一化）:\n")
        for seg_path, start, end, raw_score, norm_score in normalized_scores:
            f.write(f"{os.path.basename(seg_path)} | {start:.1f}-{end:.1f}s | 原始得分: {raw_score:.2f} | 归一化: {norm_score:.4f}\n")

        f.write("\n# 综合评分构成：\n")
        f.write(f"归一化最高美学得分: {score_highest_norm:.4f}\n")
        f.write(f"归一化最低美学得分: {score_lowest_norm:.4f}\n")
        f.write(f"场景检测阈值: {threshold_raw:.2f} → 归一化: {threshold_score_norm:.4f}\n")
        f.write(f"最小场景相似度 SSIM: {similarity_score:.4f} → 不相似度 (1 - sim): {similarity_norm:.4f}\n")

        f.write("\n# 权重：\n")
        f.write(f"w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}\n")

        f.write("\n# 综合评分计算公式:\n")
        f.write(f"score = {w1} * {score_highest_norm:.4f} + "
                f"{w2} * {score_lowest_norm:.4f} + "
                f"{w3} * {threshold_score_norm:.4f} + "
                f"{w4} * {similarity_norm:.4f}\n")

        f.write(f"\n✅ 综合美学评分（Multi-scene Aesthetic Score，归一化）: {final_score:.4f}\n")

    print(f"\n✅ 综合美学评分（归一化）: {final_score:.4f}")
    print(f"✅ 所有评分写入：{scores_file_path}")


if __name__ == '__main__':
    main()

```

```python
# @title
import os
import json

def extract_aesthetic_quality_from_results(result_dir):
    """
    遍历结果目录下所有 JSON 文件，查找其中的 "aesthetic_quality" 键，
    返回所有找到的评分列表。
    """
    aesthetic_scores = []
    # 遍历 result_dir 目录（包括子目录）
    for root, dirs, files in os.walk(result_dir):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, "r") as f:
                        data = json.load(f)
                except Exception as e:
                    print(f"读取文件 {file_path} 出错: {e}")
                    continue

                # 判断数据结构，可能是字典或列表
                if isinstance(data, dict):
                    if "aesthetic_quality" in data:
                        # 假设返回的结构是 ["aesthetic_quality": [score, details]]
                        score = data["aesthetic_quality"][0]
                        aesthetic_scores.append(score)
                elif isinstance(data, list):
                    for entry in data:
                        if isinstance(entry, dict) and "aesthetic_quality" in entry:
                            score = entry["aesthetic_quality"][0]
                            aesthetic_scores.append(score)
    return aesthetic_scores

# 示例用法
if __name__ == '__main__':
    # 请将此处 result_dir 修改为实际结果目录，例如 VBench 输出目录
    result_dir = "vbench_temp/segment_2.mp4.json"
    scores = extract_aesthetic_quality_from_results(result_dir)
    if scores:
        print("Extracted aesthetic_quality scores:", scores)
        avg_score = sum(scores) / len(scores)
        print(f"Average aesthetic_quality score: {avg_score:.4f}")
    else:
        print("未在目录中找到 'aesthetic_quality' 的相关评分。")
```
